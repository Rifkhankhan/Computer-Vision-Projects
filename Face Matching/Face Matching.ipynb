```python
import matplotlib.pyplot as plt
```

```python
# Load the known and unknown images
known_image = face_recognition.load_image_file("./../Rifkhan.jpg")
unknown_image = face_recognition.load_image_file("./../WhatsApp Image 2024-10-08 at 3.08.38 PM.jpeg")
```

```python
# Encode the faces
known_encoding = face_recognition.face_encodings(known_image)[0]
unknown_encoding = face_recognition.face_encodings(unknown_image)[0]

# Calculate the Euclidean distance between the faces
distance = face_recognition.face_distance([known_encoding], unknown_encoding)

print("Distance between faces:", distance[0])
```
