```python
import face_recognition
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the known images and encode them
known_images = ["./../Rifkhan.jpg", "./../WhatsApp Image 2024-10-08 at 3.08.55 PM.jpeg",'./../WhatsApp Image 2024-10-08 at 3.08.38 PM.jpeg']


known_encodings = [face_recognition.face_encodings(face_recognition.load_image_file(img))[0] for img in known_images]
```

below it the alternative mathode

```python
known_encodings = []
for img in known_images:
    image = face_recognition.load_image_file(img)
    encodings = face_recognition.face_encodings(image)
    if len(encodings) > 0:
        known_encodings.append(encodings[0])
    else:
        print(f"No faces found in {img}")

# Check the result
print(f"Number of known encodings: {len(known_encodings)}")
```

face_recognition.load_image_file(img): This function loads an image file from the given path (img). It reads the image and converts it into a format that can be processed by the face_recognition library.

face_recognition.face_encodings(...): This function takes the loaded image and detects the face(s) in it. It then computes the facial features (encodings) for each detected face. These encodings are numerical representations of the facial features.

[0]: Since face_recognition.face_encodings returns a list of encodings (one for each face detected in the image), [0] selects the first encoding. This assumes that there is only one face per image.

List Comprehension: The entire line is a list comprehension that iterates over each image path in known_images, loads the image, computes the face encoding, and stores the first encoding in the known_encodings list.
